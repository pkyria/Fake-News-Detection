```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from wordcloud import WordCloud
import nltk

df = pd.read_csv('LIAR.csv')

labels = df['Label']


label_val = labels.value_counts().values
label_in = labels.value_counts().index

g_pie = plt.pie(label_val,labels=label_in, autopct ='%1.1f%%')
plt.title("Label Values", bbox={'facecolor':'0.8', 'pad':5})

plt.show()


fig, ax = plt.subplots()

colors =['mistyrose','tomato','deepskyblue','lightskyblue','blue','red']

bar = ax.barh(label_in, label_val, color=colors)
ax.set_title("Label values, descending order")

plt.show()
```

```python
df['text'] = df['Subject'] + " " +df['Speaker'] + " " +df['Statement'] + " " +df['context'] #context being tested
df = df.drop(['Subject', 'Speaker', 'Statement', 'context','ID', 'Speaker Job Title', 'State Info', 'Party', 'barely true counts', 'false counts',
         'half true counts', 'mostly true counts', 'pants on fire counts'], axis=1)

df = df.dropna()

df.text = df.text.astype(str)
```

```python
text = ' '.join(text for text in df['text'])

wordcloud = WordCloud (width = 3500, height = 2500, stopwords= set(nltk.corpus.stopwords.words("english")), 
                       background_color = 'black').generate(text)

figure = plt.figure(figsize = (20,20), facecolor = "k", edgecolor = "k"
)

plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
